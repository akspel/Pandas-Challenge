```python
# Create a reference to the CSV and import it into a Pandas DataFrame
import pandas as pd 
import os
csv_path = os.path.join("Resources", "purchase_data.csv") 
purchase_df = pd.read_csv(csv_path)
purchase_df.head()
```

```python
# Player Count
```

```python
# Total number of players
# total_bugs = len(eclipse_df.index)
total_players = len(purchase_df["SN"].value_counts())

# Creating Data Fram for Total Players
player_count = pd.DataFrame({"Total Players": [total_players]})
player_count
```

```python
# Purchasing Analysis
```

```python
# Number of unique items
# Create a dataframe of the average stats for each type of pokemon.
# pokemon_group = pokemon_type_df.groupby(["Type 1"])

# pokemon_comparison_df = pokemon_group.mean()
# pokemon_comparison_df
unique_items = len((purchase_df["Item ID"]).unique())
# Average Price of Items
average_price = (purchase_df["Price"]).mean()
# Number of Purchases
number_of_purchases = (purchase_df["Purchase ID"]).count()
# Total Revenue
total_revenue = (purchase_df["Price"].sum())

# Creating data frame for Items
Item_Summary_df = pd.DataFrame({"Unqiue Items": [unique_items],
                               "Average Price": [average_price],
                               "Purchases": [number_of_purchases],
                                "Total Revenue": [total_revenue]})
Item_Summary_df
```

```python
# Gender Demographic
# Percentage and Count of Male, Female and Other/Non-Disclosed Players
gender_stats = purchase_df.groupby("Gender")

total_count_gender = gender_stats.nunique()["SN"]
total_count_gender

percentage_of_players = total_count_gender / total_players * 100
percentage_of_players

gender_demographics = pd.DataFrame({"Percentage of Players": percentage_of_players, "Total Count": total_count_gender})
gender_demographics
```

```python
# Purchasing Analysis (Gender)
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Average Purchase Total per Person by Gender

# Count the total purchases
purchase_count = gender_stats["Purchase ID"].count()

# Average purchase price
average_price = gender_stats["Price"].mean()

# Average purchase total
average_total = gender_stats["Price"].sum()

# Average purchase total per person
average_per_person = average_total/total_count_gender

# Creating data frame
gender_demographics_df = pd.DataFrame({"Purchase Count": purchase_count,
                                      "Average Purchase Price": average_price,
                                      "Average Purchae Value": average_total,
                                      "Average Purchase Total per Person": average_per_person})
gender_demographics_df.index.name = "Gender"

gender_demographics_df
```

```python
# Age Demographics
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Sort Values
# df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names, include_lowest=True)
# df
purchase_df["Age Group"] = pd.cut(purchase_df["Age"], age_bins, labels=group_names, include_lowest=True)
purchase_df.head()
```

```python
# Age Demographics
# New data frame with grouping
age_grouped = purchase_df.groupby("Age Group")

# Count total palyers by age category
total_age = age_grouped["SN"].nunique()
total_age
```

```python
# Purchase Count
# Count the total purchases
purchase_count_age = age_grouped["Purchase ID"].count()

# Average purchase price
average_price_age = age_grouped["Price"].mean()

# Average purchase total
average_total_age = age_grouped["Price"].sum()

# Average purchase total per person
average_per_person_age = average_total_age/total_age

# Creating data frame
age_demographics_df = pd.DataFrame({"Purchase Count": purchase_count_age,
                                      "Average Purchase Price": average_price_age,
                                      "Average Purchae Value": average_total_age,
                                      "Average Purchase Total per Person": average_per_person_age})
age_demographics_df.index.name = "Age"

age_demographics_df
```

```python
# Top Spenders
# Identify the top 5 spenders in the game by total purchase value
# SN
spender_stats = purchase_df.groupby("SN")
# Purchase Count
purchase_count_spender = spender_stats["Purchase ID"].count()
# Average purchase price
avg_purchase_price_spender = spender_stats["Price"].mean()
# Total purchase value
purchase_total_spender = spender_stats["Price"].sum()

# Create data frame with obtained values
top_spenders_df = pd.DataFrame({"Purchase Count": purchase_count_spender,
                            "Average Purchase Price": avg_purchase_price_spender,
                            "Total Purchase Value": purchase_total_spender})
top_spenders_df.head()
```

```python
# Most Popular Items
# Identify the 5 most popular items by purchase count
popular_items = purchase_df[["Item ID", "Item Name", "Price"]]
# Item ID
# Item Name
item_stats = popular_items.groupby(["Item ID", "Item Name"])
# Purchase Count
purchase_item_count = item_stats["Price"].count()
# Total Purchase Value
purchase_price = (item_stats["Price"].sum())
# Item Price
item_price = purchase_price/purchase_item_count

# Create data frame
most_popular_items_df = pd.DataFrame({"Purchase Count": purchase_item_count,
                                  "Item Price": item_price,
                                  "Total Purchase Value": purchase_price})
most_popular_items_df.head()
```

```python
# Most Profitable Items
# Identify the 5 most profitable items by total purchase value
profitable_items = most_popular_items_df[["Purchase Count", "Item Price", "Total Purchase Value"]]
# Item ID
# Item Name
item_stats_proft = profitable_items.groupby(["Purchase Count", "Item Price"])
# Purchase Count
profitable_item_count = profitable_items["Total Purchase Value"].count()
# Total Purchase Value
profitable_item_price = (profitable_items["Total Purchase Value"].sum())
# Item Price
profitable_price = profitable_item_price/profitable_item_count

most_profitable_items_df = pd.DataFrame({"Purchase Count": profitable_item_count,
                                         "Item Price": profitable_price,
                                        "Total Purchase Value": profitable_item_price})
most_profitable_items_df
```
